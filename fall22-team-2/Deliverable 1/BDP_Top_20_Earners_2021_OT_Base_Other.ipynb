```python


def makePlot(df, xcol, ycol, xrange, title,xlimit,xls):
    # make data
    x = xrange
    y = df[xcol]
    names = df[ycol]

    # plot
    fig, ax = plt.subplots()

    fig.set_figheight(10)
    fig.set_figwidth(10)


    cont = ax.barh(names,y, color=["#EF404E" for _ in range(2)]+['#231F20' for _ in range(len(names))])
    ax.set_facecolor('#EFECEA')
    ax.set(xlim=(0, xlimit),xticks=xls)
    plt.xlabel('Total '+title+' Earnings', fontsize=16, fontname=font)
    plt.ylabel('Boston Police Deparment Employees', fontsize=16, fontname=font)
    plt.title('Boston Police Department Top 20 '+title+' Earners 2021',fontsize=18, fontname=font)
    plt.grid(axis='x')



    ticks = ax.get_xticks()


    # formatted the values into strings beginning with dollar sign
    new_labels = [f'${amt:,.0f}.00' for amt in ticks]
    # Set the new labels
    ax.set_xticklabels(new_labels,fontname=font)
    ax.set_yticklabels(names,fontname=font)



    ax.bar_label(container=cont, labels=[f'${amt:,.0f}' for amt in y],fontsize=14,fontname=font)


    ax.invert_yaxis()
    plt.show()
```
