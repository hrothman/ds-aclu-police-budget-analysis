```python
file21=pd.read_csv('2021earn.csv',thousands=',')


df21 = pd.DataFrame(file21)
top21=df21.sort_values(by='TOTAL_GROSS',ascending=False)

top21_20=top21.head(20)
label=list(top21_20['NAME'])
value=list(top21_20['TOTAL_GROSS'])
plt.rcdefaults()
fig, ax = plt.subplots()
people=label
y_pos = np.arange(len(people))
performance=value
ax.barh(y_pos, performance)
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis()# labels read top-to-bottom
ax.set_yticklabels(people, fontsize=10)
ax.set_xlabel('unit=100')
ax.set_title('2021 Top Earners')

plt.show()
```

```python
file20=pd.read_csv('2020earn.csv',thousands=',')
df20 = pd.DataFrame(file20)
df20[' TOTAL EARNINGS '] = df20[' TOTAL EARNINGS '].replace({'\$':''}, regex = True)
for i in range(len(df20[' TOTAL EARNINGS '])):
    df20[' TOTAL EARNINGS '][i]=df20[' TOTAL EARNINGS '][i][1:-1].replace(',','')
    
df20[' TOTAL EARNINGS '] = df20[' TOTAL EARNINGS '].astype(float)
    


top20=df20.sort_values(by=' TOTAL EARNINGS ',ascending=False)

top20_20=top20.head(20)


label=list(top20_20['NAME'])
value=list(top20_20[' TOTAL EARNINGS '])

plt.rcdefaults()
fig, ax = plt.subplots()
people=label
y_pos = np.arange(len(people))
performance=value
ax.barh(y_pos, performance)
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis()# labels read top-to-bottom
ax.set_yticklabels(people, fontsize=10)

ax.set_title('2020 Top Earners')

plt.show()



```
